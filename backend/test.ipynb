```python
import numpy as np
import matplotlib.pyplot as plt

# Set the seed for reproducibility
np.random.seed(42)

# Generate synthetic data with a combination of sine and cosine functions
X = np.linspace(0, 4 * np.pi, 200)
true_curve = 2 * np.sin(X) + 1.5 * np.cos(0.5 * X)  # True curve without noise
y = true_curve + np.random.normal(0, 10, 200)  # Synthetic data with noise
y=true_curve
# Reshape X and y to match the expected format
X = X.reshape(-1, 1)  # Reshape X to [n_samples, n_features] where n_features is 1
y = y.reshape(-1)     # Reshape y to [n_samples]

# Visualize both the true curve and the synthetic data
plt.figure(figsize=(8, 6))

plt.plot(X, true_curve, label='True Curve', color='blue', linestyle='--', linewidth=2)
plt.scatter(X, y, label='Synthetic Data with Noise', color='orange', alpha=0.7)

plt.title('Synthetic Data and True Curve')
plt.xlabel('Feature (X)')
plt.ylabel('Target (y)')
plt.legend()
plt.show()
```

```python
from gplearn.genetic import SymbolicRegressor
```

```python
est_gp = SymbolicRegressor(population_size=2,
                           generations=1, stopping_criteria=0.01,
                           p_crossover=0.7, p_subtree_mutation=0.1,
                           p_hoist_mutation=0.05, p_point_mutation=0.1,
                           max_samples=0.9, verbose=1,
                           parsimony_coefficient=0.01, random_state=0,function_set=("add","sub","mul","div","sin","inv","cos"))

```

```python
population=est_gp.fit(X, y)
for i in population:
    print(i)
```

```python
est_gp.fitCalc(population)
```

```python
est_gp.set_params(generations=2, warm_start=True)
population=est_gp.fit(X, y)
for i in population:
    print(i)
```

```python
est_gp.fitCalc(population)
```

```python
est_gp = SymbolicRegressor(population_size=2,
                           generations=1, stopping_criteria=0.01,
                           p_crossover=0.7, p_subtree_mutation=0.1,
                           p_hoist_mutation=0.05, p_point_mutation=0.1,
                           max_samples=0.9, verbose=1,
                           parsimony_coefficient=0.01, random_state=0,function_set=("add","sub","mul","div","sin","inv","cos"))
for i in range(1,10):
    est_gp.set_params(generations=i, warm_start=True)
    population=est_gp.fit(X, y)
    for i in population:
        print(i)
    est_gp.fitCalc(population)
    
```
